```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.common.exceptions import NoSuchElementException

opt = Options()

opt.add_argument('--start-maximized')

driver = webdriver.Chrome(options=opt)

links = []

for i in range(40):
    # Vamos a la pagina
    driver.get(
        f'https://www.fincaraiz.com.co/apartamentos/venta/medellin?pagina={i}')

    # Obetenemos todos los elementos con links
    elementos = driver.find_elements(
        By.XPATH, '//*[@id="listingContainer"]/div/article/a')

    # obtenemos cada link
    for apartamento in elementos:
        links.append(apartamento.get_attribute('href'))

    # Removemos proyecto
    for link in links:
        if '/proyecto' in link:
            links.remove(link)

#Creamos las listas para ingestar los datos
descripcion = []
habitaciones = []
banios = []
parqueaderos = []
area_construida = []
area_privada = []
estrato = []
estado = []
antiguedad = []
piso = []
administracion = []
precio_m2 = []
barrio = []

#iteramos dentro de la pagina para obtener los datos
for link in links:
    driver.get(link)
    try: 
        descripcion.append(driver.find_element(
            By.XPATH, '//*[@id="general"]/div/div[1]/div/div/div/p[1]').text)
        barrio.append(driver.find_element(
            By.XPATH, '//*[@id="__next"]/div[2]/main/section/div[2]/div/header/div/p[2]').text)

        elementos = driver.find_element(
            By.CSS_SELECTOR, '#general > div > div.MuiBox-root.jss261 > div.MuiBox-root.jss262.jss260')
    except NoSuchElementException:
        continue
    print(elementos.text.split('\n'))

    lista = elementos.text.split('\n')

#Se crea un diccionario para luego convertirlo en dataframe

    diccionario = {}

    print(link)

    if (len(lista) % 2)!= 0:
        lista.append(None)

    for i in range(0, len(lista), 2):
        diccionario[lista[i]] = lista[i+1]

# Se agrega al diccionario los datos obtenidos de cada elemento

    habitaciones.append(diccionario.get('Habitaciones', None))
    banios.append(diccionario.get('Baños', None))
    parqueaderos.append(diccionario.get('Parqueaderos', None))
    area_construida.append(diccionario.get('Área construída', None))
    area_privada.append(diccionario.get('Área privada', None))
    estrato.append(diccionario.get('Estrato', None))
    estado.append(diccionario.get('Estado', None))
    antiguedad.append(diccionario.get('Antigüedad', None))
    piso.append(diccionario.get('Piso N°', None))
    administracion.append(diccionario.get('Administración', None))
    precio_m2.append(diccionario.get('Precio m²', None))


apartamentos = {
    'descripcion': descripcion,
    'habitaciones': habitaciones,
    'banios': banios,
    'parqueaderos': parqueaderos,
    'area_construida': area_construida,
    'area_privada': area_privada,
    'estrato': estrato,
    'estado': estado,
    'antiguedad': antiguedad,
    'piso': piso,
    'administeracion': administracion,
    'precio_m2': precio_m2,
    'barrio': barrio}
```

```python
print(apartamentos)
```

```python
max_len = max(len(v) for v in apartamentos.values())
```

```python
# Se llena con NaN los espacios vacios para mantener la dimensionalidad de los datos
for key in apartamentos:
    while len(apartamentos[key]) < max_len:
        apartamentos[key].append(None)
```

```python
# Se convierten los datos de diccionario a dataframe
apartamentos = pd.DataFrame.from_dict(apartamentos)
print(apartamentos)
```
